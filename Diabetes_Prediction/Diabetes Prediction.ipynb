```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
data = pd.read_excel("Diabetes_prediction.xlsx")

data.head()

```

```python
data.dtypes
```

```python
data['gender_id'] = data['gender'].apply(lambda x : 1 if x == 'Male' else 2)
```

```python
data.head(5)
```

```python
data['smoking_history'].unique()
```

```python
def has_smoked(sh):
    if(sh in ['never','No Info']):
        return 0
    else: 
        return 1
    
```

```python
data['has_smoked'] = data['smoking_history'].apply(lambda x : has_smoked(x))
```

```python
data.tail(10)
```

```python
# Define the input variables
X = data[['gender_id','age','bmi','blood_glucose_level','HbA1c_level','hypertension','heart_disease','has_smoked']]

# Define the target variable
y = data['diabetes']

# Train the logistic regression model
model = sm.Logit(y, X).fit()

# Extract the coefficients and intercept
coefficients = model.params[1:]
intercept = model.params[0]

# Print the coefficients and intercept
print("Logistic Regression Coefficients:", coefficients)
print("Logistic Regression Intercept:", intercept)
```
